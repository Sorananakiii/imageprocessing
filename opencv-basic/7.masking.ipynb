```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
## original image
img = cv2.imread("Data/Photos/cats.jpg")

plt.imshow(img)
plt.show()
```

```python
blank = np.zeros(img.shape[:2], dtype="uint8")
mask = cv2.circle(blank, (img.shape[1]//2, img.shape[0]//2 - 100), 150, 255, -1)
masked = cv2.bitwise_and(img, img, mask=mask)

plt.imshow(masked)
plt.show()
```
