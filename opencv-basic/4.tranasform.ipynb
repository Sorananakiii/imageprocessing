```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("Data/Photos/park.jpg")

plt.imshow(img)
plt.show()
```

```python
def translate(img, xaxis, yaxis):
    transMat = np.float32([[1, 0, xaxis], [0, 1, yaxis]])
    dim = (img.shape[1], img.shape[0])
    return cv2.warpAffine(img, transMat, dim)

# -x:left, x:right, -y:up, y:down
trans = translate(img, 100, 100)
plt.imshow(trans)
plt.show()


# -x:left, x:right, -y:up, y:down
trans = translate(img, -100, 100)
plt.imshow(trans)
plt.show()
```

```python
def rotate(img, angle, rotpoint=None):
    h, w, c = img.shape
    if rotpoint is None:
        rotpoint = (w // 2, h // 2)
    rotMat = cv2.getRotationMatrix2D(rotpoint, angle, 1.0)
    dim = (w, h)
    return cv2.warpAffine(img, rotMat, dim)

rot = rotate(img, 30)
plt.imshow(rot)
plt.show()
```

```python
# flipping
flip = cv2.flip(img, 0)

plt.imshow(flip)
plt.show()
```
