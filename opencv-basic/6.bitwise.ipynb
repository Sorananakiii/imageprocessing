```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
blank = np.zeros((400, 400), dtype="uint8")

rec = cv2.rectangle(blank.copy(), (30,30), (370, 370), 255, thickness=cv2.FILLED)
plt.imshow(rec, cmap="gray")
plt.show()

circle = cv2.circle(blank.copy(), (200,200), 200, 255, thickness=cv2.FILLED)
plt.imshow(circle, cmap="gray")
plt.show()
```

```python
## and = intersection area
bitwise_and = cv2.bitwise_and(rec, circle)
plt.imshow(bitwise_and, cmap="gray")
plt.show()
```

```python
## or = union area = non intersection and intersection area
bitwise_or = cv2.bitwise_or(rec, circle)
plt.imshow(bitwise_or, cmap="gray")
plt.show()
```

```python
bitwise_not = cv2.bitwise_not(rec)
plt.imshow(bitwise_not, cmap="gray")
plt.show()
```

```python
## xor = only non intersection
bitwise_xor = cv2.bitwise_xor(rec, circle)
plt.imshow(bitwise_xor, cmap="gray")
plt.show()
```
