```python
import cv2
import matplotlib.pyplot as plt
```

```python
## original image
img = cv2.imread("Data/Photos/cat.jpg")

plt.imshow(img)
plt.show()
```

```python
## gray scale image
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap="gray")
plt.show()
```

```python
## blur image
blur = cv2.GaussianBlur(img, (7,7), cv2.BORDER_DEFAULT)

plt.imshow(blur)
plt.show()
```

```python
## edge cascade
canny = cv2.Canny(blur, 0, 150)

plt.imshow(canny, cmap="gray")
plt.show()
```

```python
## Dilating the image
dilated = cv2.dilate(canny, (3,3), iterations=3)

plt.imshow(dilated, cmap="gray")
plt.show()
```

```python
## eroding
eroded = cv2.erode(dilated, (3,3), iterations=3)

plt.imshow(eroded, cmap="gray")
plt.show()
```

```python
nimg = cv2.resize(img, (400, 400), interpolation=cv2.INTER_AREA)

plt.imshow(nimg, cmap="gray")
plt.show()
```
