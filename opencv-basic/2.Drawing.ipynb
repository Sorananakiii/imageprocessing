```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
blank = np.zeros((500, 500, 3), dtype='uint8')

plt.imshow(blank)
plt.show()
```

```python
blank[200:300, 300:400] = 0, 0, 255

plt.imshow(blank)
plt.show()
```

```python
blank = cv2.rectangle(blank, (0, 0), (250, 250), (0, 250, 0), thickness=2)

plt.imshow(blank)
plt.show()
```

```python
dim = (blank.shape[1] // 2, blank.shape[0] // 2)
blank = cv2.rectangle(blank, (0, 0),  dim, (0, 250, 0), thickness=cv2.FILLED)

plt.imshow(blank)
plt.show()
```

```python
blank = np.zeros((500, 500, 3), dtype='uint8')
dim = (blank.shape[1] // 2, blank.shape[0] // 2)
blank = cv2.circle(blank, dim, 100, (0, 250, 0), thickness=cv2.FILLED)

plt.imshow(blank)
plt.show()
```

```python
blank = np.zeros((500, 500, 3), dtype='uint8')
dim = (blank.shape[1] // 2, blank.shape[0])
blank = cv2.line(blank, (250, 0),  dim, (0, 250, 0), thickness=1)

plt.imshow(blank)
plt.show()
```

```python
blank = np.zeros((500, 500, 3), dtype='uint8')
dim = (blank.shape[1] // 2, blank.shape[0] // 2)
blank = cv2.putText(blank, "Hello World", dim, cv2.FONT_HERSHEY_TRIPLEX, 1.0, (0,255,0), 2)

plt.imshow(blank)
plt.show()
```
